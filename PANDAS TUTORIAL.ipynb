# Filtering - Using Conditionals to Filter Rows and Columns

```python
people = {
     'first':['Muhsin','Tahseen','Rahil'],
     'last':['Bashir','Bashir','Mir'],
     'email':['muhsinbashir@gmail.com','tahseenbashir@gmail.com','rahilmir@gmail.com']
 }
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
filt = (df['last']== 'Bashir')
```

```python
df.loc[filt,'email']
```

```python
filt = (df['last']== 'Bashir') & (df['first']=='Muhsin')
```

& (and)-----     for two conditions

```python
df.loc[filt,'email']
```

```python
filt = (df['last']== 'Mir') | (df['first']=='Muhsin')
```

|(or) is used for or conditions

```python
df.loc[filt,'email']
```

```python
df.loc[~filt,'email']
```

~   is used foe getting the oppsite of that condition

# Cleaning Data - Casting Datatypes and Handling Missing Values

```python
import numpy as np
import pandas as pd
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df=pd.DataFrame(people)

df.replace('NA',np.nan,inplace=True)
df.replace('Missing',np.nan,inplace=True)
```

```python
df
```

```python
df.dropna()
```

.dropna() dropping null values

```python
df.dropna(axis='index',how='any')
```

any missing value in the row 

```python
df.dropna(axis='index',how='all')
```

all missing value in the row

```python
df.dropna(axis='columns',how='all')
```

```python
df.dropna(axis='columns',how='any')
```

```python
df.dropna(axis='index',how='any',subset=['email'])
```

if we want to delete any specific column

```python
df.dropna(axis='index',how='all',subset=['last','email'])
```

```python
df.isna()
```

```python
df.fillna(0)
```
